```python
import faker

fake = faker.Faker()

print(fake.name())
print(fake.email())
print(fake.address())
print(fake.company())
```

```python
#createma user
import json

users = []
for i in range(1000):
    user = {
        "username": fake.unique.user_name(),
        "name": fake.name(),
        "email": fake.email(),
        "address": fake.address(),
        "age": fake.random_int(min=18,max=90)
    }
    users.append(user)

with open("users.json", "w") as file:
    json.dump(users,file, indent=4)
print(users)
```

```python
#crate an ad campaign
#satart and end dates
#target user age demographic
#currency

import random
from datetime import date, timedelta

def get_start_end_dates():
    duration = random.randint(1,2*365)
    offset = random.randint(-365,365)
    start = date.today() - timedelta(days=offset)
    end = start + timedelta(days=duration)

    return start.strftime("%Y%m%d"), end.strftime("%Y%m%d")

print(get_start_end_dates())

#target demographic
def get_age_range():
    age = random.randrange(20, 46, 5)
    diff = random.randrange(5, 21, 5)

    return f"{age}-{age + diff}"

print(get_age_range())

def get_currency():
    return random.choice(("GBP", "USD", "EUR"))

def get_compaign_name():
    start, end = get_start_end_dates()
    name = start + '_' + end + '_'
    name += get_age_range() + '_'
    name += get_currency()
    return name

print(get_compaign_name())
```

```python
def get_compaign_data():
    name = get_compaign_name()
    budget = random.randint(10**3,10**6)
    spent = random.randint(10**2,budget)

    return {
        "cmp_name": name,
        "cmp_budget": budget,
        "cmp_spent": spent
    }
print(get_compaign_data())

def get_raw_data(users):
    raw_data = []
    for user in users:
        campaigns = []
        for i in range(random.randint(5,10)):
            campaign_data = get_compaign_data()
            age_range = campaign_data["cmp_name"].split("_") [2].split("-")
            min_age = int(age_range[0])
            max_age = int(age_range[1])
            if user["age"] >= min_age and user["age"] <= max_age:
                campaigns.append(campaign_data)

        raw_data.append({"user": user, "campaigns": campaigns})
    return raw_data

raw_data = get_raw_data(users)

with open("raw_data.json", "w") as file:
    json.dump(raw_data, file, indent = 4)


campaign_data = []
for data in raw_data:
    for campaign in data["campaigns"]:
        campaign["user"] = data["user"]
        campaign_data.append(campaign)

with open("campaign_data.json", "w") as file:
    json.dump(campaign_data, file, indent = 4)
```
